```python
import urllib.request as rq
import pandas as pd
import numpy as np
```

***We are importing a file for italy's daywise covid records and an overall global record of all countries.***

```python
url = 'https://gist.githubusercontent.com/aakashns/f6a004fa20c84fec53262f9a8bfee775/raw/f309558b1cf5103424cef58e2ecb8704dcd4d74c/italy-covid-daywise.csv'
path = 'italy_coviddata'
rq.urlretrieve(url, path)
df  = pd.read_csv(path)
df

```

```python
url2 = 'https://gist.githubusercontent.com/aakashns/8684589ef4f266116cdce023377fc9c8/raw/99ce3826b2a9d1e6d0bde7e9e559fc8b6e9ac88b/locations.csv'
path2 = 'locations.csv'
rq.urlretrieve(url2, path2)
df2 = pd.read_csv(path2)
```

- **FOLLOWING OPERATIONS ARE DONE IN THE CELL BELOW:**<br>
    1. ***Converting records in negative numbers to positive numbers.***<br>
    2. ***Setting print options to display maximum rows.***<br>
    3. ***Converting the datatype of the date column in italy dataset to datetime.***
    4. ***Removing any NaNs from the data and replacing it with 0.0***<br>

```python

df.at[176, "new_cases"] = (df.at[177, "new_cases"] + df.at[178, "new_cases"] + df.at[175, "new_cases"] + df.at[174, "new_cases"])/2.0
l1 = ["new_cases", "new_tests", "new_deaths"]
for i in l1:
    for j in range(248):
        if (df.at[j, i] ) <= -1:
            df.at[j,i]= df.at[j,i]* (-1)
            
pd.set_option('display.max_rows', 250)

df["date"] = pd.to_datetime(df.date)

df.fillna(0.0, inplace=True)


df["positivity_rate"] = 0.0
df["death_rate"] = 0.0
for i in l1:
    for j in range(248):
        if df.at[j,"new_cases"] == 0.0 or df.at[j,"new_cases"] == "NaN" or df.at[j,"new_tests"] == 0.0 or df.at[j,"new_cases"] == "NaN" or df.at[j,"new_deaths"] == 0.0 or df.at[j,"new_deaths"] == "NaN":
            df.at[j, "positivity_rate"] = 0.0
            df.at[j, "death_rate"] = 0.0
        else:
            df.at[j,"positivity_rate"] = df.at[j, "new_cases"]/ df.at[j, "new_tests"]
            df.at[j, "death_rate"] = df.at[j, "new_deaths"]/df.at[j, "new_cases"]
```

```python
initial_tests = 935310
df["total_cases"] = df.new_cases.cumsum()
df["total_deaths"] = df.new_deaths.cumsum()
df["total_tests"] = df.new_tests.cumsum()
```

```python
df["location"] = "Italy"

df['year'] = pd.DatetimeIndex(df.date).year
df['month'] = pd.DatetimeIndex(df.date).month
df['day'] = pd.DatetimeIndex(df.date).day
df['weekday'] = pd.DatetimeIndex(df.date).weekday
covid_month_df=df.groupby("month")[['new_cases', 'new_deaths', 'new_tests']].sum()
covid_month_df



```

```python
merged_df = df.merge(df2, on = "location")
```

```python
merged_df['cases_per_million'] = merged_df.total_cases * 1e6 / merged_df.population

merged_df['deaths_per_million'] = merged_df.total_deaths * 1e6 / merged_df.population

merged_df['tests_per_million'] = merged_df.total_tests * 1e6 / merged_df.population

result_df = merged_df[['date',
                       'new_cases', 
                       'total_cases', 
                       'new_deaths', 
                       'total_deaths', 
                       'new_tests', 
                       'total_tests', 
                       'cases_per_million', 
                       'deaths_per_million', 
                       'tests_per_million']]

```
